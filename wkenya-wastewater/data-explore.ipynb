## Vis. W Kenya Data

````python
import pandas as pd
import geopandas as gpd
````

````python
hw_file = "data/HydroWASTE_v10.csv"
````

````python
df = pd.read_csv(hw_file, encoding="ISO-8859-1")
df.shape
````

````python
ken_df = df[df["CNTRY_ISO"]=="KEN"].copy()
ken_df.shape
````

````python
ken_df
````

````python
gdf = gpd.GeoDataFrame(ken_df, geometry=gpd.GeoSeries.from_xy(df["LON_WWTP"], df["LAT_WWTP"]), crs=4326)
gdf.shape
````

````python
gdf.explore?
````

````python
gdf.explore(width="75%", height="100%", tiles="CartoDB positron", popup=True)
````
